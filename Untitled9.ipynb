```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
df=pd.read_csv('/content/student_cheating_dataset-1.csv')
df
```

```python
average_score = df['score'].mean()
print(f"The average score is: {average_score:.2f}")
```

```python
# Bar chart of cheating vs non-cheating students
sns.countplot(x='cheated', data=df)
plt.title('Count of Cheating vs. Non-Cheating Students')
plt.xlabel('Cheated (1: Yes, 0: No)')
plt.ylabel('Count')
plt.xticks([0, 1], ['No', 'Yes'])
plt.show()
```

```python
# Bar chart of cheating method counts
sns.countplot(x='cheating_method', data=df)
plt.title('Count of Cheating Methods')
plt.xlabel('Cheating Method')
plt.ylabel('Count')
plt.xticks(rotation=45, ha='right')
plt.tight_layout()
plt.show()
```

```python
# Scatter plot of score vs. exam duration for cheating and non-cheating students
sns.scatterplot(x='exam_duration_minutes', y='score', hue='cheated', data=df)
plt.title('Score vs. Exam Duration by Cheating Status')
plt.xlabel('Exam Duration (minutes)')
plt.ylabel('Score')
plt.show()
```

```python
# Heatmap of correlation between numerical columns
numerical_cols = df.select_dtypes(include=np.number).columns
sns.heatmap(df[numerical_cols].corr(), annot=True, cmap='coolwarm', fmt=".2f")
plt.title('Correlation Matrix of Numerical Columns')
plt.show()
```

```python

# Insights and Observations:
print("\n--- Insights and Observations ---")
print(f"The average score across all exams is approximately {average_score:.2f}.")
print(f"The dataset contains {df['cheated'].sum()} instances of cheating.")
print("The bar chart shows the distribution of cheating and non-cheating students.")
print("The cheating method bar chart indicates the frequency of different cheating methods used.")
print("The scatter plot suggests a potential relationship between exam duration, score, and cheating status.")
print("The heatmap reveals the correlation between various numerical features in the dataset.")
```
